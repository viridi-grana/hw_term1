#### Задача 9 (\*) - 2 балла. 

Пользователь вводит количество минут. Это число - время до выхода Миши на концерт (на сборы и поездку). Задача - напечатать действия (по одной строке в результате проверки условий), которые Миша точно успеет сделать (с учетом времени, которое требуется на дорогу). Действия: Принять ванну (2 часа 10 минут), порепетировать (1 час 20 минут), поменять струны (15 минут), побриться (7 минут). Поездка до клуба займет 1 час 5 минут. Если Мише не хватает времени ни на одно дело, выведите "Миша поедет неподготовленным". Если Мише не хватает времени даже на дорогу, выведите "Миша опоздал на концерт".

```python
time = int(input())
if time >= 130 + 80 + 15 + 7 + 65:
  print('Миша успеет принять ванну, порепетировать, поменять струны и побриться')
elif time >= 80 + 15 + 7 + 65:
  print('Миша успеет порепетировать, поменять струны и побриться')
elif time >= 15 + 7 + 65:
  print('Миша успеет поменять струны и побриться')
elif time >= 15 + 65:
  print('Миша успеет поменять струны или побриться')
elif time >= 7 + 65:
  print('Миша успеет побриться')
elif time >= 65:
  print('Миша поедет неподготовленным')
else:
  print('Миша опоздал на концерт')
```

#### Задача 10 (\*) - 2 балла.

Пользователь вводит, сколько часов по времени в Москве (счет времени 24). Ивана срочно вызвали в командировку, и Маша не успела даже спросить, куда он едет. Спустя некоторое время от него приходит сообщение: "У нас час ночи, я пошел спать". Маша смотрит на часы и видит введенное пользователем количество часов. Маша догадывается, что Ивана могли отправить в столицу одной из этих стран: Америка, Япония, Китай, Бразилия или Испания. Помогите Маше узнать, в какой стране сейчас Иван. Если ее догадки насчет возможных стран ложны, выведите "Маша не знает, где Иван".

```python
hour_moscow = int(input())
if hour_moscow == 25 - 6:
  print ('Иван в Японии')
elif hour_moscow == 25 - 5:
  print ('Иван в Китае')
elif hour_moscow == 1 + 1:
  print ('Иван в Испании')
elif hour_moscow == 1 + 6:
  print ('Иван в Бразилии')
elif hour_moscow == 1 + 7:
  print ('Иван в США')
else:
   print ('Маша не знает, где Иван')
```
